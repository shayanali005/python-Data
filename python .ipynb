# NetworkX

```python
import networkx as nx
```

```python
G =nx.Graph()
G.add_nodes_from(["Ali","Hussain","Mehdi"])
G.add_edges_from([("Ali","Hussain"),("Hussain","Mehdi")])
nx.draw(G,with_labels = "True", node_color = "orange", node_size = 1000, font_weight = "bold")
plt.show()
```

```python
G = nx.Graph()
G.add_edges_from([("Muhammad", "Ali"), ("Ali", "Fatima"), ("Fatima", "Hassan"), ("Hassan", "Hussain"),("Hussain","Muhammad")])

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos, node_color='red', node_size=1000)
nx.draw_networkx_edges(G, pos, width=2, edge_color='yellow')
nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif")

plt.title("Customized Social Network")
plt.show()
```
